```python
class calculator():# Class is made for calculating the age
    def __init__ (self):
        pass
    
    def month(self):# method for calculating age in months
        cal = self.age * 12
        print(f"\nAge Calculated in Months is {cal} Months")
        
    def week(self): # method for calculating age in weeks
        cal = self.age * 52
        print(f"\nAge Calculated in Weeks is {cal} Weeks")
        
    def day(self): # method for calculating age in days
        cal = self.age * 365
        print(f"\nAge Calculated in Days is {cal} Days")
        
    def hour(self): # method for calculating age in hours
        cal = self.age * 365 * 24
        print(f"\nAge Calculated in Hours is {cal} Hours")
    
    def minute(self): # method for calculating age in minutes
        cal = self.age * 365 * 24 * 60
        print(f"\nAge Calculated in Minutes is {cal} Minutes")
    
    def second(self): # method for calculating age in seconds
        cal = self.age * 365 * 24 * 60 * 60
        print(f"\nAge Calculated in Seconds is {cal} Seconds")
        

def age_calculator(person): # Function for perfoming the age calculation
    
    print("*********** Welcome To The Age Calculator ************\n")
    find = int(input("\nPlease enter the age to be calculated in integers: "))# input from user to calculate age
    
    if find > 0: # condition check if age is more than zero
        person.age = find
        
    else:
        print("\nWarning, Please enter the age more than zero")
        print("\nDo you want to do another calculation?")
        cont = int(input("\nPress 1 for Yes or Press any number for No "))
        if cont == 1:
            age_calculator(person)
        else:
            print("\nThank you For Using Age Calculator")
            return
            
    print("\nType one of the Keywords to calculate age accordingly:") # input for calculating age as per requirement
    operation = input("\nm-->Months, w-->Weeks, d-->Days, h-->Hours, n-->Minutes, s-->Seconds ").lower()
    
    if operation == "m": # m for calculation in months
        person.month()
        
        print("\nDo you want to do another calculation?")
        cont = int(input("\nPress 1 for Yes or Press any number for No "))
        if cont == 1:
            age_calculator(person)
        else:
            print("\nThank you For Using Age Calculator")
    
    elif operation == "w": # w for calculation in weeks
        person.week()
        
        print("\nDo you want to do another calculation?")
        cont = int(input("\nPress 1 for Yes or Press any number for No "))
        if cont == 1:
            age_calculator(person)
        else:
            print("\nThank you For Using Age Calculator")
    
    elif operation == "d": # d for calculation in days
        person.day()
        
        print("\nDo you want to do another calculation?")
        cont = int(input("\nPress 1 for Yes or Press any number for No "))
        if cont == 1:
            age_calculator(person)
        else:
            print("\nThank you For Using Age Calculator")
    
    elif operation == "h": # h for calculation in hours
        person.hour()
        
        print("\nDo you want to do another calculation?")
        cont = int(input("\nPress 1 for Yes or Press any number for No "))
        if cont == 1:
            age_calculator(person)
        else:
            print("\nThank you For Using Age Calculator")
    
    elif operation == "n": # n for calculation in minutes
        person.minute()
        
        print("\nDo you want to do another calculation?")
        cont = int(input("\nPress 1 for Yes or Press any number for No "))
        if cont == 1:
            age_calculator(person)
        else:
            print("\nThank you For Using Age Calculator")
    
    elif operation == "s": # s for calculation in seconds
        person.second()
        
        print("\nDo you want to do another calculation?")
        cont = int(input("\nPress 1 for Yes or Press any number for No "))
        if cont == 1:
            age_calculator(person)
        else:
            print("\nThank you For Using Age Calculator")
    
    else:
        print("\nInput is not acceptable, calculator closed")
        print("\nDo you want to do another calculation?")
        cont = int(input("\nPress 1 for Yes or Press any number for No "))
        if cont == 1:
            age_calculator(person)
        else:
            print("\nThank you For Using Age Calculator")

if __name__=="__main__":
    
    person = calculator() # calling the class and making the boject person
    age_calculator(person) # calling the function to rum the codes
```
